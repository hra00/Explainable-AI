# Visualization : Cam EigenCam GradCam

```python
from cams.visualizer import Visualizer
from utils.image import get_img_list_raf, img_to_input_tensor, pp_images, pp_blended_heatmaps, aligned_tensor
from utils.labels import get_labels
import numpy as np
```

## Parameters

```python
MODEL_NAME = 'RAF-impr-std_0124-1008_weights.h5'
LAST_CONV_NAME = 'block3_conv3'
IMG_SHAPE = (100,100,3)
emotion_labels = get_labels(MODEL_NAME)
print(emotion_labels)
n_imgs = {'happiness':18, 'surprise':21, 'sadness':18, 'anger':20, 'disgust':20, 'fear':20}
```

## Visualizer

```python
vis = Visualizer(MODEL_NAME, LAST_CONV_NAME, IMG_SHAPE)
```

## get heatmaps : Cam, EigenCam, GradCam

```python
imgs = []
heatmaps_C, heatmaps_C_l = [], []
heatmaps_E = []
heatmaps_G, heatmaps_G_l = [], []
preds = []

def mean_imgs(imgs:list):
    mean_i = np.mean(np.array(imgs), axis=0)
    return mean_i

for emotion in emotion_labels[:-1]:
    correct_label = emotion_labels.index(emotion)
    num_imgs=n_imgs[emotion]
    img_list = ['./data/emotion/aligned/{emotion}/{emotion}_{n}.png'.format(n=i, emotion=emotion) for i in range(1,num_imgs+1)]
    img_tensors, img_tensors_rgb, img_tensors_not_rgb = img_to_input_tensor(img_list, IMG_SHAPE)

    heatmap_CAM, pred_CAM = vis.getCAM(img_tensors_not_rgb, preprocess=True)
    heatmap_CAM_l, pred_CAM_l = vis.getCAM(img_tensors_not_rgb, preprocess=True, start_idx=correct_label)
    heatmap_EigenCam, pred_EigenCam = vis.getEigenCAM(img_tensors_not_rgb, preprocess=True)
    heatmap_GradCam, pred_GradCam = vis.getGradCAM(img_tensors_not_rgb, preprocess=True)
    heatmap_GradCam_l, pred_GradCam_l = vis.getGradCAM(img_tensors_not_rgb, preprocess=True, start_idx=correct_label)

    imgs.append(mean_imgs(img_tensors))
    heatmaps_C.append(mean_imgs(heatmap_CAM))
    heatmaps_C_l.append(mean_imgs(heatmap_CAM_l))
    heatmaps_E.append(mean_imgs(heatmap_EigenCam))
    heatmaps_G.append(mean_imgs(heatmap_GradCam))
    heatmaps_G_l.append(mean_imgs(heatmap_GradCam_l))

    preds.append(pred_CAM)
```

```python
def get_accuracy(preds, correct_label):
    mean_sicherheit = np.array([float(pred) for idx, pred in preds if idx == correct_label]).mean()
    accuracy = sum(1 for idx, _ in preds if idx == correct_label) / len(preds)
    return mean_sicherheit.__round__(2), accuracy.__round__(2)

accuracy = [get_accuracy(pred_emotion, correct_label) for pred_emotion, correct_label in zip(preds, [0,1,2,3,4,5])]
```

## Result

```python
methods = ['CAM', 'CAM (correct)', 'EigenCAM','GradCAM','GradCAM (correct)']
labels = [e + ' ' + str(acc) for e, acc in zip(emotion_labels[:-1], accuracy)]
[labels.extend([methode + ' : ' + emotion for emotion in emotion_labels[:-1]]) for methode in methods]

pp_images(imgs*6, imgs+heatmaps_C+heatmaps_C_l+heatmaps_E+heatmaps_G+heatmaps_G_l, labels = labels, wd=6 , figsize=(20,20), fontsize=60)
```
